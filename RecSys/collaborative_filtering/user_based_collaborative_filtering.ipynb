```python
def pcc(u: pd.core.series.Series, v: pd.core.series.Series) -> float:
    mean_u = 0.0
    mean_v = 0.0
    
    flag = True
    for idx, item in u.items():
        if pd.isna(u[idx]) or pd.isna(v[idx]):
            continue        
        mean_u += u[idx]
        mean_v += v[idx]
        flag=False
        
    if flag:
        return 0.0
    
    
    mean_u /= u.shape[0]
    mean_v /= v.shape[0]
    
    std_u =0.0
    std_v =0.0
    deno = 0.0
    
    for idx, item in u.items():
        if pd.isna(u[idx]) or pd.isna(v[idx]):
            continue
        deno += (u[idx]-mean_u)*(v[idx]-mean_v)
        std_u += math.pow((u[idx]-mean_u),2)
        std_v += math.pow((v[idx]-mean_v),2)
        
    std_u = math.sqrt(std_u)
    std_v = math.sqrt(std_v)
    

    return deno/std_u/std_v
        
        
        
        
```
